```python
def allplots(num):
    Species = (data[:, [num]]).tolist()
    empty_list = []
    for sublist in Species:
        for item in sublist:
            empty_list.append(item)
    Species = empty_list
    y = Species
    N = len( y )
    x = range( N )
    width = 1/1.5
    plt.bar( x, y, width, color="blue" )
    plt.show()
```

```python
allplots(6)
```

```python
allplots(8)
```

```python
import os
datapath = "/home/eebc177student/Developer/repos/eeb-c177-project/analyses"
workingdir = "/home/eebc177student/Developer/repos/eeb-c177-project/analyses/sorted_by_species"
```

```python
#made a file that inputs a species name and outputs all the sites that species has interactions in
sp = (input("Name of species: "))
os.chdir(datapath)
with open("Karan_Dataset.csv") as fr:
        reader = csv.DictReader(fr)
        header = reader.fieldnames
        os.chdir(workingdir)
        with open("{}.csv".format(sp), "w") as fw:
            writer = csv.DictWriter(fw, fieldnames = header, delimiter = ",")
            os.chdir(datapath)
            for row in reader:
                if row["species"] == "{}".format(sp):
                    writer.writerow(row)
```
